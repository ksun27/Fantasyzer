```python
def get_player_name():
    with open('../nba_player_data_20-21.csv', encoding='utf-8') as f:
        reader = csv.reader(f, delimiter=",", quotechar='"')
        next(reader, None)
        data_read = [row for row in reader]
        for i in range(len(data_read)):
            #if data_read[i][1] == None:
            yield data_read[i][1]
```

```python
for value in get_player_name(): 
    print(value)
```

```python
import pandas as pd
df = pd.read_csv('../nba_player_data_20-21.csv')
def removeDash(x):
    if '-' in x:
        x = x[x.index('-'):]
        x = x.replace("-", "")
        return x
    else:
        return x
df['Pos'] = df['Pos'].apply(lambda x: removeDash(x))
print(df)
#df.to_csv('consolidated_players.csv')
```
